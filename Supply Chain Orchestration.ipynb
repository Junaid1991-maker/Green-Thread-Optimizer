```python
#The "Loom-to-Linen" Orchestrator
import sqlite3
import pandas as pd
import hashlib
import hmac
import secrets
from datetime import datetime

# 1. Initialize Secure In-Memory Database
# Using ':memory:' ensures no footprint on your SSD
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

def initialize_erp():
    # Table 1: US Retail Inventory (Demand Side)
    cursor.execute('''CREATE TABLE US_Inventory 
                     (SKU TEXT PRIMARY KEY, Stock INTEGER, Safety_Stock INTEGER, Unit_Price_USD REAL)''')
    
    # Table 2: Karachi Production Floor (Supply Side)
    # Includes 'Energy_Window' to simulate Karachi's K-Electric Peak/Off-Peak cycles
    cursor.execute('''CREATE TABLE Karachi_Production 
                     (Loom_ID TEXT PRIMARY KEY, Status TEXT, Energy_Window TEXT, Capacity_Daily INTEGER)''')

    # Seed Data - Reliable, industry-standard values
    inventory_data = [('TOWEL-G7-ELITE', 420, 1200, 15.50)]
    production_data = [('LOOM-KHI-01', 'IDLE', 'PEAK_TARIFF', 2500)]
    
    # Using parameterized queries to prevent SQL Injection
    cursor.executemany("INSERT INTO US_Inventory VALUES (?, ?, ?, ?)", inventory_data)
    cursor.executemany("INSERT INTO Karachi_Production VALUES (?, ?, ?, ?)", production_data)
    conn.commit()
    print("✅ ERP System Initialized: Secure & In-Memory")

initialize_erp()
```

```python
#This script initializes a secure, in-memory SQLite database with two relational tables designed to bridge US demand and Karachi production. 
It uses parameterized queries and a memory-only footprint to ensure data integrity, speed, and cybersecurity.
```

```python
# --- MOCKED JAVA ENTERPRISE SERVICE (Logic-Only) ---
# This dictionary simulates a strictly-typed Spring Boot Microservice
class JavaSecurityHub:
    def __init__(self):
        # A secret key for HMAC (In production, this would be in an AWS Secret Manager)
        self._secret_key = b"Karachi-US-Secure-Bridge-2024"
        self.version = "v2.6-Enterprise"

    def validate_order_logic(self, gsm, quantity):
        """Strict validation logic mimicking Java's type safety."""
        # GSM (Grams per Square Meter) must be within export quality standards
        is_quality_valid = 300 <= gsm <= 900
        is_qty_valid = 0 < quantity <= 10000
        return is_quality_valid and is_qty_valid

    def generate_digital_handshake(self, payload):
        """Generates a cryptographic signature for the Loom Controller."""
        signature = hmac.new(self._secret_key, payload.encode(), hashlib.sha256)
        return signature.hexdigest().upper()

# Instantiate our "Bridge"
java_service = JavaSecurityHub()
```

```python
#This class simulates a high-integrity Java backend that validates production parameters against export standards and secures data using HMAC-SHA256 
cryptography. It acts as a digital "firewall" to ensure that only authorized and quality-checked instructions are sent to the loom.
```

```python
#The Agentic Orchestrator (The "Brains")
def run_supply_chain_orchestration():
    print(f"--- 🛰️  Executing Orchestration: {datetime.now().strftime('%Y-%m-%d %H:%M')} ---")
    
    # 1. Check for Stockouts (Parameterized Query)
    target_sku = 'TOWEL-G7-ELITE'
    cursor.execute("SELECT Stock, Safety_Stock FROM US_Inventory WHERE SKU=?", (target_sku,))
    stock, safety = cursor.fetchone()
    
    if stock < safety:
        deficit = safety - stock
        order_qty = deficit * 2  # Economic Order Quantity logic
        print(f"⚠️  ALERT: {target_sku} below safety levels ({stock}/{safety}).")
        
        # 2. Check Karachi Industrial Status
        cursor.execute("SELECT Loom_ID, Energy_Window FROM Karachi_Production WHERE Status='IDLE'")
        loom_id, energy_status = cursor.fetchone()
        
        # 3. SRE Logic: Energy Cost Optimization
        # Avoiding Karachi's Peak Tariff (usually 6PM-10PM)
        scheduled_time = "IMMEDIATE"
        energy_savings_flag = False
        
        if energy_status == 'PEAK_TARIFF':
            print(f"🕒 LOCAL REALITY: K-Electric Peak Hours detected for {loom_id}.")
            print("🔄 RESCHEDULING: Shifting production to OFF-PEAK window to save ~18% cost.")
            scheduled_time = "23:00 PKT (Night Shift)"
            energy_savings_flag = True

        # 4. Java-Style Validation & Cryptographic Signing
        export_gsm = 550
        if java_service.validate_order_logic(export_gsm, order_qty):
            # Create a data string for the loom to verify
            order_manifest = f"SKU:{target_sku}|QTY:{order_qty}|START:{scheduled_time}"
            token = java_service.generate_digital_handshake(order_manifest)
            
            return {
                "Status": "✅ PRODUCTION_SECURED",
                "Assigned_Loom": loom_id,
                "Batch_Quantity": order_qty,
                "Execution_Window": scheduled_time,
                "Security_Hash": token[:16], # Only show first 16 for logs
                "ROI_Optimization": "K-Electric Tariff Avoidance" if energy_savings_flag else "Standard"
            }
    
    return {"Status": "Inventory Levels Nominal"}

# Execute the simulation
report = run_supply_chain_orchestration()
print("\n--- 📝 FINAL ARCHITECT REPORT ---")
for key, value in report.items():
    print(f"{key}: {value}")
```

```python
#The "Load Shedding" Resilience Update
import sqlite3
from datetime import datetime

# Connect to our existing in-memory DB
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

def initialize_resilient_erp():
    # US Side: Added a 'Marketing_Status' to show how production affects sales
    cursor.execute('''CREATE TABLE US_Demand_Center 
                     (SKU TEXT, Stock INTEGER, Marketing_Status TEXT)''')
    
    # Karachi Side: Added Grid_Status and Generator_Fuel_Liters
    cursor.execute('''CREATE TABLE Karachi_Industrial_Zone 
                     (Loom_ID TEXT, Status TEXT, Grid_Status TEXT, Gen_Fuel_Liters INTEGER)''')

    # Seed Data: Grid is OFFLINE, and Fuel is CRITICALLY LOW (only 40L)
    cursor.execute("INSERT INTO US_Demand_Center VALUES ('TOWEL-G7-ELITE', 420, 'ACTIVE_ADS')")
    cursor.execute("INSERT INTO Karachi_Industrial_Zone VALUES ('LOOM-KHI-01', 'IDLE', 'OFFLINE', 40)")
    conn.commit()

initialize_resilient_erp()

def resilient_orchestrator():
    print(f"--- ⚡ SRE Fault-Tolerance Check: {datetime.now().strftime('%H:%M')} ---")
    
    # 1. Check Karachi Grid Status
    cursor.execute("SELECT Loom_ID, Grid_Status, Gen_Fuel_Liters FROM Karachi_Industrial_Zone")
    loom_id, grid, fuel = cursor.fetchone()
    
    # 2. Fault Logic: Load Shedding Handler
    if grid == 'OFFLINE':
        print(f"⚠️  CRITICAL: {loom_id} reports Grid Power is DOWN (Load Shedding).")
        
        # Check if we can failover to the Generator
        if fuel < 100:  # Threshold: 100 Liters needed for a full shift
            print(f"🚫 FAILOVER FAILED: Generator Fuel is too low ({fuel}L). Production Aborted.")
            
            # 3. Graceful Degradation: Notify US Marketing to stop spending money
            cursor.execute("UPDATE US_Demand_Center SET Marketing_Status = 'PAUSED' WHERE SKU='TOWEL-G7-ELITE'")
            conn.commit()
            
            return {
                "Action": "STOP_PRODUCTION",
                "Reason": "Grid Offline + Low Fuel",
                "US_Countermeasure": "Marketing Ads PAUSED (Save Budget)",
                "Risk_Mitigation": "Prevented 'Out-of-Stock' Customer Dissatisfaction"
            }
        else:
            print(f"🔋 FAILOVER SUCCESS: Switching to Backup Generator ({fuel}L available).")
            return {"Action": "PROCEED_ON_GENERATOR"}

    return {"Action": "PROCEED_ON_GRID"}

# Execute the Fault-Tolerance Simulation
fault_report = resilient_orchestrator()

print("\n--- 🛡️  SRE INCIDENT REPORT ---")
for key, value in fault_report.items():
    print(f"{key}: {value}")
```

```python
#The "Monte Carlo" Risk Engine
import random
import sqlite3
import numpy as np # Standard for data science, though we can use pure Python

# 1. Setup the Simulation Environment
current_stock = 420
safety_stock_threshold = 1200
mean_daily_sales = 50   # Average towels sold per day
std_dev_sales = 25      # Volatility (how much demand swings)
days_to_forecast = 30   # Planning for the next month

def run_monte_carlo_forecast(iterations=1000):
    print(f"--- 🎲 Running Monte Carlo Simulation ({iterations} paths) ---")
    
    stockout_count = 0
    simulated_final_stocks = []

    for _ in range(iterations):
        # Simulate 30 days of sales using a Normal Distribution (Gaussian)
        # random.gauss(mean, standard_deviation)
        monthly_demand = sum([max(0, random.gauss(mean_daily_sales, std_dev_sales)) 
                             for _ in range(days_to_forecast)])
        
        final_stock = current_stock - monthly_demand
        simulated_final_stocks.append(final_stock)
        
        # If stock drops below 0, it's a Stockout incident
        if final_stock < 0:
            stockout_count += 1

    # Calculate Probability
    prob_stockout = (stockout_count / iterations) * 100
    avg_predicted_stock = sum(simulated_final_stocks) / iterations

    print(f"📊 Average Predicted Stock: {int(avg_predicted_stock)} units")
    print(f"🚨 Probability of Stockout: {prob_stockout:.2f}%")

    # 2. Agentic Decision Logic
    if prob_stockout > 15.0:
        print("\n--- 🧶 Karachi Procurement Agent: HIGH RISK DETECTED ---")
        print("⚡ ACTION: Pre-ordering 5,000kg of Raw Yarn in Karachi to lock in current prices.")
        return {
            "Risk_Level": "HIGH",
            "Stockout_Probability": f"{prob_stockout}%",
            "Mitigation": "Bulk Yarn Purchase Triggered",
            "Financial_Logic": "Hedge against price spikes during peak demand"
        }
    
    return {"Risk_Level": "LOW", "Action": "Maintain Standard Production"}

# Run the Predictive Analytics
risk_report = run_monte_carlo_forecast()

print("\n--- 📈 PREDICTIVE RISK REPORT ---")
for key, value in risk_report.items():
    print(f"{key}: {value}")
```

```python
#The "Green-Loom" Carbon Optimizer
import sqlite3
from datetime import datetime

# 1. Setup ESG Database Layer
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

def initialize_esg_erp():
    # Emissions factors (kg of CO2 per kWh)
    # Karachi Grid is roughly 0.45kg/kWh (Gas/Coal) | Solar is 0.05kg/kWh (Lifecycle)
    cursor.execute('''CREATE TABLE Energy_Sources 
                     (Source_Name TEXT, Carbon_Intensity REAL, Cost_per_kWh REAL)''')
    
    # US Carbon Policy
    cursor.execute('''CREATE TABLE US_Regulatory_Env 
                     (Region TEXT, Carbon_Tax_Per_KG REAL, Penalty_Threshold_KG REAL)''')

    cursor.execute("INSERT INTO Energy_Sources VALUES ('NATIONAL_GRID', 0.45, 0.12)")
    cursor.execute("INSERT INTO Energy_Sources VALUES ('OFF_PEAK_GRID', 0.45, 0.09)")
    cursor.execute("INSERT INTO Energy_Sources VALUES ('PRIVATE_SOLAR', 0.05, 0.00)") # Sun is free!
    
    # Simulation: US imposes a high tax of $0.10 per kg of CO2 emitted
    cursor.execute("INSERT INTO US_Regulatory_Env VALUES ('US_EAST', 0.10, 500.0)")
    conn.commit()

initialize_esg_erp()

def esg_optimized_orchestrator(batch_units=1000):
    print(f"--- 🌍 ESG Sustainability Audit: {datetime.now().strftime('%Y-%m-%d')} ---")
    
    energy_required_kwh = (batch_units / 100) * 5  # Estimated 5kWh per 100 towels
    
    # 1. Fetch Energy and Regulatory Data
    cursor.execute("SELECT Source_Name, Carbon_Intensity, Cost_per_kWh FROM Energy_Sources")
    sources = cursor.fetchall() # List of (Name, CO2, Cost)
    
    cursor.execute("SELECT Carbon_Tax_Per_KG FROM US_Regulatory_Env WHERE Region='US_EAST'")
    carbon_tax_rate = cursor.fetchone()[0]

    best_plan = None
    min_total_cost = float('inf')

    # 2. Multi-Objective Optimization: Calculate (Energy Cost + Carbon Tax)
    for name, co2_rate, energy_price in sources:
        energy_cost = energy_required_kwh * energy_price
        total_co2 = energy_required_kwh * co2_rate
        carbon_tax_cost = total_co2 * carbon_tax_rate
        
        total_landed_cost = energy_cost + carbon_tax_cost
        
        print(f"Checked {name}: Cost ${energy_cost:.2f} | CO2 {total_co2:.1f}kg | Tax ${carbon_tax_cost:.2f}")

        if total_landed_cost < min_total_cost:
            min_total_cost = total_landed_cost
            best_plan = {
                "Source": name,
                "Total_Cost_USD": round(total_landed_cost, 2),
                "Carbon_Footprint": f"{total_co2:.1f} kg CO2",
                "Sustainability_Index": "High" if name == 'PRIVATE_SOLAR' else "Low"
            }

    # 3. Decision Logic
    if best_plan["Source"] == 'PRIVATE_SOLAR':
        execution_note = "☀️ SOLAR PRIORITY: Scheduling production for 11:00 AM PKT to utilize peak sunlight."
    else:
        execution_note = "🌙 GRID PRIORITY: Scheduling for Night Shift to minimize energy tariff."

    print(f"\n{execution_note}")
    return best_plan

# Execute the ESG Simulation
esg_report = esg_optimized_orchestrator()

print("\n--- 🧾 SUSTAINABILITY & COST REPORT ---")
for key, value in esg_report.items():
    print(f"{key}: {value}")
```

```python
#The "Heartbeat" & System Integrity Monitor
import time
import sys
from datetime import datetime

# 1. The Watchdog System
def run_integrity_heartbeat(service_instance, retries=3):
    print(f"--- 🛡️  SRE Watchdog: Monitoring JavaSecurityHub ---")
    
    # We simulate a "Real-Time" loop (3 pings for this portfolio demo)
    for i in range(1, retries + 1):
        timestamp = datetime.now().strftime('%H:%M:%S')
        
        try:
            # PING: Checking if the object exists and has the correct attributes
            # In SRE, we check for 'Liveness' and 'Readiness'
            status_code = 200 if hasattr(service_instance, 'version') else 500
            
            # INTEGRITY CHECK: Ensure the version hasn't been downgraded or tampered with
            if status_code == 200 and service_instance.version == "v2.6-Enterprise":
                print(f"[{timestamp}] 🟢 PING {i}: Hub Healthy (Version: {service_instance.version})")
            else:
                raise ValueError("Integrity Breach: Security Version Mismatch")
                
        except (NameError, AttributeError, ValueError) as e:
            print(f"[{timestamp}] 🔴 PING {i}: {str(e)}")
            trigger_system_panic()
            return False
            
        time.sleep(1) # Simulating a 1-second interval heartbeat
        
    print("✅ Heartbeat Stable: All Loom Orders UNLOCKED.")
    return True

def trigger_system_panic():
    print("\n" + "!"*45)
    print("🚨 !!! SYSTEM PANIC: SECURITY HUB OFFLINE !!! 🚨")
    print("🔒 ACTION: GLOBAL LOOM LOCKDOWN INITIATED.")
    print("📡 ALERT: Sending SMS/Email to SRE Lead in Karachi.")
    print("!"*45 + "\n")
    # In a real SRE scenario, this would trigger a sys.exit() or a circuit breaker
    return "LOCKDOWN_ACTIVE"

# --- SCENARIO 1: Healthy System ---
print("TESTING SCENARIO: NORMAL OPERATION")
run_integrity_heartbeat(java_service)

# --- SCENARIO 2: Tampered/Deleted System ---
print("\nTESTING SCENARIO: SECURITY BREACH")
# Simulating a "hacker" modifying the version or deleting the object
java_service.version = "v1.0-Hacked" 
run_integrity_heartbeat(java_service)
```
